# implementation K-Means

## import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

from sklearn.datasets import load_iris

warnings.filterwarnings('ignore')
```

## Data preprocessing

```python
data = load_iris()
X = pd.DataFrame(data.data, columns = data.feature_names)
y = pd.DataFrame(data.target, columns = ['Species'])
df = pd.concat([X, y], axis = 1)
df.head()
```

```python
df = df.loc[:, ['sepal length (cm)', 'petal length (cm)', 'Species']]
df.head()
```

```python
X = df[['sepal length (cm)', 'petal length (cm)']].values
```

```python
fig, ax = plt.subplots()

ax.scatter(df.loc[df['Species'].isin([0]), 'sepal length (cm)'], df.loc[df['Species'].isin([0]), 'petal length (cm)'], color = 'red', marker = 'o', label = 'Label1')
ax.scatter(df.loc[df['Species'].isin([1]), 'sepal length (cm)'], df.loc[df['Species'].isin([1]), 'petal length (cm)'], color = 'blue', marker = 'o', label = 'Label2')
ax.scatter(df.loc[df['Species'].isin([2]), 'sepal length (cm)'], df.loc[df['Species'].isin([2]), 'petal length (cm)'], color = 'green', marker = 'o', label = 'Label3')

plt.xlabel('sepal length (cm)')
plt.ylabel('petal length (cm)')
plt.show()
```

## K-Means algorithm

```python
centroids = df.sample(n=3).values
```

```python
def kmeans(X, k, iteration): #kmeans 함수
    cluster = np.zeros(X.shape[0]) #0으로 초기화된 150차원(data point 갯수: 150) vector space 생성
    global centroids
    for i in range(iteration): #iteration 횟수만큼 반복
        for i, row in enumerate(X): #i refers index, row refers sepal, petal length
            mn_dist = float('inf') #greedy algorithm 사용을 위해 float('inf') 선언
            for idx, centroid in enumerate(centroids): #center 좌표 구함
                d = np.sqrt((centroid[0]-row[0])**2 + (centroid[1]-row[1])**2) #유클리드 거리 계산
                if mn_dist > d:
                    mn_dist = d #mn_dist에 계산한 거리 저장
                    cluster[i] = idx #i 번째 클러스터에 학습 결과 저장
        centroids = pd.DataFrame(X).groupby(by=cluster).mean().values
    return centroids, cluster
```

## Run K-Means

```python
centroids, cluster = kmeans(X, 3, 1)
```

```python
sns.scatterplot(X[:, 0], X[:, 1], hue=cluster)
sns.scatterplot(centroids[:, 0], centroids[:, 1], s=100, color='y')
plt.xlabel('sepal length (cm)')
plt.ylabel('petal length (cm)')
plt.show()
```
